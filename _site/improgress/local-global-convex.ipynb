# Intuitive guide to the how Local Minimser are Global Minimiser if the constrain anf function are convex

While learning about non-linear optimisation,My professor mentioned one of the most important theorem in non-linear optimsiation, Local minima are global minima theorem. I was facinated by if we premise formally defined as, 

$\qquad$ Theorem : Let $\Omega \in R^n$ be a open convex set and $f : \Omega \rightarrow R$. Assume $x^*$ is a local minimser of $f$ on $\Omega \Rightarrow x^*$ is a global minimser of $f$ on $\Omega$

In this case $\Omega$ is the feasible region which is a set of all the constrains and f is the objective function of the optimisation problem, so, 

$$
\min_{\Omega} \quad \Omega \quad \text{subject to} \quad f
$$

Interestingly, this means for instance i have a convex feaisble region and convex objective function then does this mean if the feaisble region only has one minimum point since the local minumum = global minimum, i mean since it's not strict then the function could have multiple global minumums. So if i find the local minumum then its a global minimum?

An examples of this is say we have a defined, 

$$\qquad f(x_1, x_2)= x_{1}^2 + x_{2}^2$$ 

$$\qquad \Omega = \{x \in R^2 | -1 \leq x_1 \leq 1, -1 \leq x_2 \leq 1\}.$$

Visually this means say we have and objective function defined as,


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the function
def f(x1, x2):
    return x1**2 + x2**2

# Generate x1 and x2 values
x1 = np.linspace(-5, 5, 100)
x2 = np.linspace(-5, 5, 100)
X1, X2 = np.meshgrid(x1, x2)
Z = f(X1, X2)

# Plot the function
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X1, X2, Z, cmap='viridis')

# Set labels and title
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('f(x1, x2)')
ax.set_title('Graph of f(x1, x2) = x1^2 + x2^2')

# Show plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x1, x2):
    return x1**2 + x2**2

# Generate x1 and x2 values
x1 = np.linspace(-2, 2, 100)
x2 = np.linspace(-2, 2, 100)
X1, X2 = np.meshgrid(x1, x2)
Z = f(X1, X2)

# Plot the function
plt.figure()
plt.contourf(X1, X2, Z, cmap='viridis')

# Add Omega as a box
plt.fill_between([-1, 1], -1, 1, color='red', alpha=0.3)

# Set labels and title
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Contour Plot of f(x1, x2) = x1^2 + x2^2 with Omega')

# Show plot
plt.colorbar()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x1, x2):
    return x1**2 + x2**2

# Generate x1 and x2 values
x1 = np.linspace(-12, 0, 100)
x2 = np.linspace(-12, 0, 100)
X1, X2 = np.meshgrid(x1, x2)
Z = f(X1, X2)

# Plot the function
plt.figure()
plt.contourf(X1, X2, Z, cmap='viridis')

# Add Omega as a box
plt.fill_between([-10, -9], -10, -9, color='red', alpha=0.3)

# Set labels and title
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Contour Plot of f(x1, x2) = x1^2 + x2^2 with Omega')

# Show plot
plt.colorbar()
plt.grid(True)
plt.show()
```

Cleary as we can in this case, f only has one global minimimser to say if we found a local minimser on $\Omega$ then this would corespond to a global minimser of $f$ on $\Omega$ for not only for this fesible region but also for f since in this case. Now say if a move the $$\Omega = \{x \in R^2 | -3 \leq x_1 \leq -2, -3 \leq x_2 \leq 2\}.$$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the function
def f(x1, x2):
    return x1**2 + x2**2

# Generate x1 and x2 values
x1 = np.linspace(-5, 5, 100)
x2 = np.linspace(-5, 5, 100)
X1, X2 = np.meshgrid(x1, x2)
Z = f(X1, X2)

# Plot the function
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X1, X2, Z, cmap='viridis')

# Set labels and title
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('f(x1, x2)')
ax.set_title('Graph of f(x1, x2) = x1^2 + x2^2')

# Add Omega as a box
x1_box = [-3, -2, -2, -3, -3]
x2_box = [-3, -3, 2, 2, -3]
z_box = [0, 0, 0, 0, 0]  # Z coordinate for the box

# Plot wireframe box
ax.plot(x1_box, x2_box, z_box, color='r')

# Set plot limits for zooming out
ax.set_xlim([-5, 5])
ax.set_ylim([-5, 5])

# Show plot
plt.show()

```
